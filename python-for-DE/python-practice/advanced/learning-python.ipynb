```python
 # pivot tables

data = {
    "A": ["foo", "bar", "bar", "foo", "foo", "bar"],
    "B": ["one", "one", "two", "two", "one", "two"],
    "C": ["x", "y", "z", "y", "z", "x"],
    "D": [1, 2, 3, 4, 5, 6]
}

df = pd.DataFrame(data)
df
```

```python
pivot_df = df.pivot_table(
    values="D",
    index="A",
    columns=["C"],
    aggfunc="sum"
)

pivot_df
```

```python
pivot_df.reset_index()
```
